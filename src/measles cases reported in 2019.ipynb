```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
measles_imported_data= "../Resources/Measles_Imported_YTD-Week_Ending_13APR2019_CDC.csv"
measles_indigenous_data= "../Resources/Measles_Indigenous_YTD-Week_Ending_13APR2019_CDC.csv"
```

```python
#measles 2019 outbreak
measles_imp = pd.read_csv(measles_imported_data)
measles_ind = pd.read_csv(measles_indigenous_data)
combined_measles = pd.merge(measles_imp, measles_ind,
                                how='outer', on='Location')
combined_measles= combined_measles.rename(columns={"Location": "State/Territory",
                                                         "Case Count_x":"Indigenous",
                                                         "Case Count_y":"Imported"})
combined_measles.head()
cols = ['Indigenous', 'Imported']
combined_measles[cols] = combined_measles[cols].apply(pd.to_numeric, errors='coerce', axis=1)
combined_measles
new_data = combined_measles.dropna(subset = ["Indigenous", "Imported"], how = "all")
new_data


#charting

states_and_territories = new_data["State/Territory"]
x_axis = np.arange(len(new_data))
plt.figure(figsize=(22,10))
tick_locations = [value+0.4 for value in x_axis]
plt.bar(x_axis, new_data["Indigenous"], color='r', alpha=0.5, align="center")
plt.bar(x_axis, new_data["Imported"], color='blue', alpha=0.5, align="edge")
plt.xticks(tick_locations, states_and_territories, rotation="vertical")
plt.title("2019 Measles Cases in the US - Year to Date")
plt.xlabel("State")
plt.ylabel("Number of Cases")

my_y_ticks = np.arange(0,130,10)
plt.yticks(my_y_ticks)
plt.savefig("../Images/2019 Measles Cases in the US.png")
```
